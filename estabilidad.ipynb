```python
def f(x):
    if 0 <= x and x<=0.5:
        return 2 * x
    elif 0.5 <= x and x <= 1:
        return 2 - (2 * x)
    
x = np.arange(0,1,0.001)
figura = plt.plot(x, [f(y) for y in x])
figura = plt.plot(x, x)
```
